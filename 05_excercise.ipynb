##### 1. The table shows the market capitalisation of the four companies. Market capitalisation is the value of a company and is calculated by multiplying the number of shares with the share price.

```python
#(a) Produce a pie chart for the four companies.

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

company = [["Barclay", 33367],
           ["BP", 68785],
           ["Tesco", 20979],
           ["Vodafone", 29741]]

df_company = pd.DataFrame(data=company, columns=("Company", "Market cap"))
print(df_company)

plt.figure()
plt.pie(df_company["Market cap"], labels=df_company["Company"],autopct='%1.1f%%')
plt.title("Four companies")
plt.show()
```

```python
#(b) Create a pie chart showing the market
plt.figure()
plt.pie(df_company["Market cap"], labels=df_company["Company"],autopct='%1.1f%%')
plt.title("Four companies")
plt.show()
```

##### 2. Read in the annual returns from BCS_ann.csv, BP_ann.csv, TSC_ann.csv, VOD_ann.csv and plot them as a box plot. Label the plot with the ticker symbols, i.e. BCS and so on.

```python
import os
print(os.getcwd())
import pandas as pd

df_bcs = pd.read_csv('BCS_ann.csv')
df_bp = pd.read_csv('BP_ann.csv')
df_tsc = pd.read_csv('TSC_ann.csv')
df_vod = pd.read_csv('VOD_ann.csv')
```

```python
plt.figure()
plt.boxplot([df_bcs["price"],df_bp["price"],df_tsc["price"],df_vod["price"]])
plt.ylabel("price")
plt.show()
```
